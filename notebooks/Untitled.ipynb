```python
from pathlib import Path
```

```python
files = Path("../examples/iu_filter/")
```

```python
files = list(files.glob("*"))
```

```python
files
```

```python
for file in files:
    b, num = file.stem.split('_')
    print(f"{b}_{num.zfill(5)}")
    print(file.parent)
    file.rename(Path('../examples/iu_filter', f"{b}_{num.zfill(5)}.jpg"))
```
